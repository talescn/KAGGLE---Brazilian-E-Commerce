# Brazilian E-Commerce Public Dataset by Olis


```python
import kagglehub
import pandas as pd
import os

# Download latest version
path = kagglehub.dataset_download("olistbr/brazilian-ecommerce")

print("Path to dataset files:", path)
```

```python
print("Arquivos no dataset:", os.listdir(path))
```

## olist_orders_dataset.csv

```python
# Escolher o arquivo CSV (por exemplo, "olist_orders_dataset.csv")
orders = os.path.join(path, "olist_orders_dataset.csv")

# Carregar no Pandas
df = pd.read_csv(orders)

df
```

```python
status = df["order_status"].value_counts()
status = status.drop("delivered")
status
```

```python
import matplotlib.pyplot as plt

# Criar o grÃ¡fico de barras
plt.figure(figsize=(10, 6))
status.plot(kind='bar')

# ConfiguraÃ§Ãµes do grÃ¡fico
plt.title('Order Status Counts')
plt.xlabel('Order Status')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')

# Ajustar o grid para ficar menos visÃ­vel
plt.grid(which='both', linestyle='--', linewidth=0.5, alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
df.info()
```

```python
df["order_purchase_timestamp"] = pd.to_datetime(df["order_purchase_timestamp"])
df["order_approved_at"] = pd.to_datetime(df["order_approved_at"])
df["order_delivered_carrier_date"] = pd.to_datetime(df["order_delivered_carrier_date"])
df["order_delivered_customer_date"] = pd.to_datetime(df["order_delivered_customer_date"])
df["order_estimated_delivery_date"] = pd.to_datetime(df["order_estimated_delivery_date"])
```

```python
df.info()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='order_status', y='order_purchase_timestamp', data=df)
plt.xticks(rotation=45)


plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Criar uma nova coluna com o tempo desde a compra atÃ© a aprovaÃ§Ã£o (em horas)
df["time_to_approve"] = (df["order_approved_at"] - df["order_purchase_timestamp"]).dt.total_seconds() / 3600  # Convertendo para horas

# Criar o grÃ¡fico boxplot comparando o tempo por status do pedido
plt.figure(figsize=(10, 6))
sns.boxplot(x='order_status', y='time_to_approve', data=df)
plt.xticks(rotation=45)
plt.ylabel("Tempo para aprovaÃ§Ã£o (horas)")
plt.title("Tempo para AprovaÃ§Ã£o do Pedido por Status")
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.show()
```

```python
import numpy as np

# Definir um limite baseado no percentil 99 para remover outliers
limite_superior = np.percentile(df["time_to_approve"].dropna(), 99)

# Filtrar valores extremos
df_filtrado = df[df["time_to_approve"] <= limite_superior]

# Criar o grÃ¡fico novamente
plt.figure(figsize=(10, 6))
sns.boxplot(x="order_status", y="time_to_approve", data=df_filtrado)
plt.xticks(rotation=45)
plt.ylabel("Tempo para aprovaÃ§Ã£o (horas)")
plt.title("Tempo para AprovaÃ§Ã£o do Pedido por Status (Sem Outliers)")
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.show()
```
