```python
import matplotlib.pyplot as plt
import seaborn as sns

# Criar uma nova coluna com o tempo desde a compra atÃ© a aprovaÃ§Ã£o (em horas)
df["time_to_approve"] = (df["order_approved_at"] - df["order_purchase_timestamp"]).dt.total_seconds() / 3600  # Convertendo para horas

# Criar o grÃ¡fico boxplot comparando o tempo por status do pedido
plt.figure(figsize=(10, 6))
sns.boxplot(x='order_status', y='time_to_approve', data=df)
plt.xticks(rotation=45)
plt.ylabel("Tempo para aprovaÃ§Ã£o (horas)")
plt.title("Tempo para AprovaÃ§Ã£o do Pedido por Status")
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.show()
```

```python
import numpy as np

# Definir um limite baseado no percentil 99 para remover outliers
limite_superior = np.percentile(df["time_to_approve"].dropna(), 99)

# Filtrar valores extremos
df_filtrado = df[df["time_to_approve"] <= limite_superior]

# Criar o grÃ¡fico novamente
plt.figure(figsize=(10, 6))
sns.boxplot(x="order_status", y="time_to_approve", data=df_filtrado)
plt.xticks(rotation=45)
plt.ylabel("Tempo para aprovaÃ§Ã£o (horas)")
plt.title("Tempo para AprovaÃ§Ã£o do Pedido por Status (Sem Outliers)")
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.show()
```

```python
df = df.dropna(subset=['atraso_entrega'])
```

```python
print(df['atraso_entrega'].dtype)
```

```python
q1 = np.percentile(df['atraso_entrega'], 1)
q99 = np.percentile(df['atraso_entrega'], 99)

print(f"Percentil 1%: {q1}")
print(f"Percentil 99%: {q99}")
```

```python
q1 = np.percentile(df['atraso_entrega'], 25)
q3 = np.percentile(df['atraso_entrega'], 75)
iqr = q3 - q1

limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

print(f"Limite inferior: {limite_inferior}")
print(f"Limite superior: {limite_superior}")

# Valores considerados extremos
extremos = df[(df['atraso_entrega'] < limite_inferior) | (df['atraso_entrega'] > limite_superior)]
print(extremos)
```

```python
# Criando um novo DataFrame sem outliers
df_sem_outliers = df[(df['atraso_entrega'] >= limite_inferior) & (df['atraso_entrega'] <= limite_superior)]

# Exibindo as primeiras linhas para conferir
print(df_sem_outliers.head())
```
